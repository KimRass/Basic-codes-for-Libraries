```python
driver = webdriver.Chrome(executable_path = "chromedriver.exe")
delay = 10

for theater_index in movie_latlondf.index:
    
    for apt_index in aptdf.index:
        # 아파트단지위치 겟
        url0 = 'https://map.naver.com/v5/search?c=14185061.3951766,4388337.5673965,12,0,0,0,dh'
        driver.get(url0)
        
        searchkey0 = aptdf.loc[apt_index]['lat_lon']
        
        textFields0 = driver.find_element_by_class_name("input_search") # 이거안됌
        textFields0.send_keys(searchkey0)
        
        driver.find_element_by_class_name("input_search").send_keys(Keys.ENTER)
        
        WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.XPATH, '//*[@id="container"]/div[1]/shrinkable-layout/entry-layout/entry-address/div/div[2]/div/div[1]/div[2]/div[1]')))
        
        apt_address = driver.find_element_by_class_name("summary_title").text
        
        
        # 영화관위치 겟
        url1 = 'https://map.naver.com/v5/search?c=14185061.3951766,4388337.5673965,12,0,0,0,dh'
        driver.get(url1)
        
        searchkey1 = movie_latlondf.loc[theater_index]['lat_lon']
        
        textFields1 = driver.find_element_by_class_name("input_search")
        textFields1.send_keys(searchkey1)
                
        driver.find_element_by_class_name("input_search").send_keys(Keys.ENTER)
        
        WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.XPATH, '//*[@id="container"]/div[1]/shrinkable-layout/entry-layout/entry-address/div/div[2]/div/div[1]/div[2]/div[1]')))
        
        theater_address = driver.find_element_by_class_name("summary_title").text
        
        
        ### 위의 두 위치를 이용, 여기서부터 길찾기 시작
        url = 'https://map.naver.com/v5/directions/-/-/-/car?c=14133144.8150121,4512932.8806473,15,0,0,0,dh'
        driver.get(url)

        searchkey2 = apt_address
        searchkey3 = theater_address          
        
        WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.XPATH, '//*[@id="directionStart0"]')))
        
        textFields2 = driver.find_element_by_xpath('//*[@id="directionStart0"]')
        textFields2.send_keys(searchkey2)
        driver.find_element_by_class_name("input_search").send_keys(Keys.ENTER)
        
        WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.XPATH, '//*[@id="container"]/div[1]/shrinkable-layout/directions-layout/directions-result/directions-search-list/search-list/search-list-contents/perfect-scrollbar/div/div[1]/div/div/div/search-item-address/div/div[2]')))     
        
        textFields3 = driver.find_element_by_xpath('//*[@id="directionGoal1"]')
        textFields3.send_keys(searchkey3)
        driver.find_element_by_xpath('//*[@id="directionGoal1"]').send_keys(Keys.ENTER)
        
        WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.XPATH, '//*[@id="container"]/div[1]/shrinkable-layout/directions-layout/directions-result/directions-search-list/search-list/search-list-contents/perfect-scrollbar/div/div[1]/div/div/div/search-item-address/div/div[2]/div/span')))
        
        driver.find_element_by_xpath('//*[@id="container"]/div[1]/shrinkable-layout/directions-layout/directions-result/div[1]/directions-search/div[2]/button[3]').click()

        WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.XPATH, '//*[@id="container"]/div[1]/shrinkable-layout/directions-layout/directions-result/directions-summary-list/div[2]/div/div/directions-summary-item-car[1]')))
        
        dist = driver.find_element_by_xpath('//*[@id="container"]/div[1]/shrinkable-layout/directions-layout/directions-result/directions-summary-list/div[2]/div/div/directions-summary-item-car[1]/div[1]/span[2]').text
        print(aptdf.loc[apt_index, 'name'], dist)
        aptdf.loc[apt_index, 'dist'] = dist
```

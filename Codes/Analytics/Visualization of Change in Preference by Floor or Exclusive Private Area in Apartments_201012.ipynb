```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import font_manager,rc
from matplotlib.pylab import rcParams

font_path = "C:/Windows/Fonts/gulim.ttc"
font_name = mpl.font_manager.FontProperties(fname=font_path).get_name()
mpl.rc("font", family=font_name)
```

## 층 선호도 변화 분석

### 시기별 변화

```python
floor = pd.read_csv("D:/☆디지털혁신팀/☆실거래가 분석/☆데이터/☆마스터 데이터/층 선호도 변화 분석_201012.csv", encoding="euc-kr", index_col=0)
cases_floor = pd.read_csv("D:/☆디지털혁신팀/☆실거래가 분석/☆데이터/☆마스터 데이터/층별 분양가 사례_201012.csv", index_col=0)
```

```python
fig, ax = plt.subplots(figsize=(14, 8))

ax.plot(floor.index, floor["층_전국_1908~2007"], linewidth=1.5, color="black", label="현재(19.08 ~ 20.07) 전국 실거래가")
ax.plot(floor.index, floor["층_전국_1408~1507"], linewidth=1.5, color="black", linestyle='--', label="5년전(14.08 ~ 15.07) 전국 실거래가")
ax.plot(floor.index, floor["층_전국_0908~1007"], linewidth=1.5, color="black", linestyle='-.', label="10년전('09.08 ~ 10.07) 전국 실거래가")

ax.set_xticks(floor.index)
ax.set_yticks(np.arange(0, 1.31, 0.05))
ax.tick_params(axis="x", labelrotation=90, labelsize=14)
ax.tick_params(axis="y", labelsize=14)
ax.axis([1, 29, 1, 1.25])
ax.grid(True, axis="both", color="black", alpha=0.3, linestyle="--")

ax.set_xlabel("층", fontsize=18)
ax.set_ylabel("전용면적당 거래금액의 비율", fontsize=18)

ax.legend(fontsize= 15, loc="upper left")

fig.tight_layout()

fig.savefig("D:/☆디지털혁신팀/☆실거래가 분석/☆그래프/floor_201012.png", bbox_inches="tight")
```

## 전용면적 선호도 변화 분석

### 시기별 변화

```python
area = pd.read_csv("D:/☆디지털혁신팀/☆실거래가 분석/☆데이터/☆마스터 데이터/전용면적 선호도 변화 분석_201012.csv", encoding="euc-kr", index_col=0)
cases_area = pd.read_excel("D:/☆디지털혁신팀/☆실거래가 분석/☆데이터/☆마스터 데이터/전용면적별 분양가 사례_201012.xlsx")
```

```python
fig, ax = plt.subplots(figsize=(14, 8))

ax.plot(area["area_range"], area["전용면적_전국_1908~2007"], linewidth=1.5, color="black", label="현재(19.08 ~ 20.07) 전국 실거래가")
ax.plot(area["area_range"], area["전용면적_전국_1408~1507"], linewidth=1.5, color="black", linestyle='--', label="5년전(14.08 ~ 15.07) 전국 실거래가")
ax.plot(area["area_range"], area["전용면적_전국_0908~1007"], linewidth=1.5, color="black", linestyle='-.', label="10년전(09.08 ~ 10.07) 전국 실거래가")

ax.set_xticks(area["area_range"])
ax.set_yticks(np.arange(0, 1.55, 0.1))
ax.tick_params(axis="x", labelrotation=90, labelsize=14)
ax.tick_params(axis="y", labelsize=14)
ax.axis([0, 24, 0.5, 1.3])
ax.grid(True, axis="both", color="black", alpha=0.3, linestyle="--")

ax.set_xlabel("전용면적(㎡)", fontsize=18)
ax.set_ylabel("전용면적당 거래금액의 비율", fontsize=18)

ax.legend(fontsize= 15, loc="upper left")

fig.tight_layout()

fig.savefig("D:/☆디지털혁신팀/☆실거래가 분석/☆그래프/area_201012.png", bbox_inches="tight")
```

### 분양가 사례와의 비교

```python
names = ["삼성물산 래미안 엘리니티", "현대건설 힐스테이트 부평", "롯데건설 백양산 롯데캐슬 골드센트럴", "GS건설 주안파크자이 더 플래티넘", "금호산업 다사역 금호어울림 센트럴", "금호산업 여주역 금호어울림 베르티스", "롯데건설 르엘 신반포", "롯데건설 의정부 롯데캐슬 골드포레",  "현대산업개발 광주 계림 IPARK SK VIEW", "현대산업개발 당진 IPARK"]
colors = ["royalblue", "orange", "lawngreen", "red", "cornflowerblue", "cornflowerblue", "cornflowerblue", "cornflowerblue", "cornflowerblue", "cornflowerblue"]

fig, ax = plt.subplots(figsize=(16, 8))

ax.plot(area.index, area["전용면적_전국_1908~2007"], linewidth=1.5, color="black", label="201908~202007 전국 실거래가")

for i, name in enumerate(names):
    if i in [0, 1, 2]:
        color = colors[i]
        print(name, color)
        df = cases_area[cases_area["단지명"]==name]
        
        sb.scatterplot(ax=ax, x=df["index"], y=df["mean"], s=30, color=color, label=name, edgecolor="black", linewidth=0.5)

ax.tick_params(axis="x", labelrotation=90, labelsize=14)
ax.tick_params(axis="y", labelrotation=0, labelsize=14)
ax.set(xticks=area.index, yticks=np.arange(0, 1.6, 0.1))
ax.set_xticklabels(area["area_range"])
ax.axis([1, 22, 0.5, 1.2])

ax.set_xlabel("면적 구간", fontsize=16)
ax.set_ylabel("전용면적당 가격의 비율", fontsize=16)

ax.grid(which="both", axis="both", linestyle="--")

ax.legend(fontsize= 14, loc="upper left")

for tick in ax.xaxis.get_major_ticks():
    tick.label.set_fontsize(13) 

fig.tight_layout()
    
fig.savefig("D:/☆디지털혁신팀/☆실거래가 분석/☆그래프/area_분양가_201012.png", bbox_inches="tight")
```
